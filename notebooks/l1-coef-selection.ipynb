```python
import matplotlib.pyplot as plt
plt.rc('font', family='Arial')
plt.rc('font', size=20)
import numpy as np
import torch
from   models import DPCCA
from   data import GTExV8Config

cfg = GTExV8Config()
```

```python
def get_l1_norm(l1_coef):
    directory = '/Users/gwg/dmcm/experiments/20190128_big_sweep/'\
                'gtexv8_mode-dpcca_batch_size-128_seed-0_latent_dim-2_lr-0.0001_l1_coef-%s_private_z-1_linear_gene_enc-0_clip-1' % l1_coef
    model_path = directory + '/model.pt'
    state = torch.load(model_path, map_location={'cuda:0': 'cpu'})
    model = DPCCA(cfg, latent_dim=2, private_z=True, linear_gene_enc=False)
    model.load_state_dict(state)
    
    n_sparse = 0
    n_params = 0
    with torch.no_grad():
        l1_norm = 0
        for param in model.pcca.parameters_('y2'):
            n_params += np.prod(param.shape)
            n_sparse += (torch.abs(param) < 0.00001).sum()
            l1_norm  += torch.norm(param, 1)
    pct_sparse = n_sparse.numpy() / n_params
    
    with open(directory + '/out.txt') as f:
        index = 0
        for i, line in enumerate(f):
            if line.strip() == 'Training model.':
                index = i
                break
                
        curve = []
        lines = [l.strip() for l in f][index:]
        train_err_ind = 2
        test_err_ind = 7
        for l in lines:
            try:
                parts = l.split('\t')
                curve.append(float(parts[train_err_ind]))
            except:
                pass
    
    return l1_norm.item(), pct_sparse, curve
```

```python
xs  = [0.0, 0.5, 1.0]
ys1 = []
ys2 = []
ys3 = []

for x in xs:
    l1_norm, pct_sparse, curve = get_l1_norm(x)
    ys1.append(l1_norm)
    ys2.append(pct_sparse)
    ys3.append(curve)

xs_ = []
for x in xs:
    xs_.append(float(x))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
plt.scatter(xs_, ys1, color='red', s=200)
plt.ylabel('L1 norm')
plt.xlabel('L1 coef')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
plt.scatter(xs_, ys2, color='blue', s=200)
plt.ylabel('% sparse')
plt.xlabel('L1 coef')
```

```python
plt.clf()
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
for l1_coef, curve in zip(xs, ys3):
    plt.plot(range(len(curve)), np.array(curve) / np.array(curve).size, label=l1_coef)
plt.legend()
plt.show()
```
