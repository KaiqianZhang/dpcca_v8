```python
import numpy as np
import torch
import random
import matplotlib
import matplotlib.pyplot as plt
from   PIL import Image
```

```python
directory = '/Users/gwg/dmcm/experiments/20190128_big_sweep/'\
            'gtexv8_mode-dpcca_batch_size-128_seed-0_latent_dim-2_lr-0.0001_'\
            'l1_coef-1.0_private_z-1_linear_gene_enc-%s_clip-1'

fname_lin = (directory % '1') + '/out.txt'
fname_non = (directory % '0') + '/out.txt'
```

```python
def get_curve(fname):
    
    with open(fname) as out:
        lines = [l.strip() for l in out]

    idx = None
    for i, l in enumerate(lines):
        if l == 'Training model.':
            idx = i+2

    losses = lines[idx:]
    img_curve = []
    gen_curve = []
    nll_curve = []
    for l in losses:
        try:
            parts = l.split('\t')
            # train indices: 2, 3, 5
            # test indices : 7, 8, 10
            img_curve.append(float(parts[7]))
            gen_curve.append(float(parts[8]))
            nll_curve.append(float(parts[10]))
        except:
            pass
            
    return img_curve, gen_curve, nll_curve

img_curve_lin, gen_curve_lin, nll_curve_lin = get_curve(fname_lin)
img_curve_non, gen_curve_non, nll_curve_non = get_curve(fname_non)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

x = list(range(len(img_curve_lin)))
y = img_curve_lin
plt.plot(x, y, color='blue', label='images - linear')
x = list(range(len(img_curve_non)))
y = img_curve_non
plt.plot(x, y, color='red', label='images - nonlinear')
plt.legend()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

x = list(range(len(img_curve_lin)))
y = gen_curve_lin
plt.plot(x, y, color='blue', label='genes - linear')
x = list(range(len(img_curve_non)))
y = gen_curve_non
plt.plot(x, y, color='red', label='genes - nonlinear')
plt.legend()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

x = list(range(len(nll_curve_lin)))
y = nll_curve_lin
plt.plot(x, y, color='blue', label='nll - linear')
x = list(range(len(nll_curve_non)))
y = nll_curve_non
plt.plot(x, y, color='red', label='nll - nonlinear')
plt.legend()
```
