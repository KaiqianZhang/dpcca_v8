# CCA-VAE

This notebook explores how a **variational loss** plus a **dissimilarity penalty** between unpaired samples of various modalities effects the results.

Model:

- A **linear model** for the gene expression data.
- A **variational autoencoder** for the images.

### Notation

Define the dataset $\mathcal{D}$ as:

$$
\mathcal{D} = \{(x_i, w_i)\}_{i=1}^n
$$

Where $x_i$ is the $i$-th image, and $w_i$ is the $i$-th gene expression levels.

---

### Experiments

- Conv CCA + VAE + dissimilar images:

    $$
    \min \sum_{i=1, j \neq i}^{n} \Big\{ \text{cosDist}(f_x(x_i), f_w(w_i)) - \color{#ff0000}{\text{cosDist}(f_x(x_j), f_w(w_i))} \Big\}
    $$

- Conv CCA + VAE + dissimilar genes:

    $$
    \min \sum_{i=1, j \neq i}^{n} \Big\{ \text{cosDist}(f_x(x_i), f_w(w_i)) - \color{#0000ff}{\text{cosDist}(f_x(x_i), f_w(w_j))} \Big\}
    $$

- Conv CCA + VAE + dissimilar images & genes:

    $$
    \min \sum_{i=1, j \neq i}^{n} \Big\{ \text{cosDist}(f_x(x_i), f_w(w_i)) - \color{#ff0000}{\text{cosDist}(f_x(x_j), f_w(w_i))} - \color{#0000ff}{\text{cosDist}(f_x(x_i), f_w(w_j))} \Big\}
    $$
