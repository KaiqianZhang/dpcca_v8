```python
import numpy as np
import torch
import random
import matplotlib
import matplotlib.pyplot as plt
from   PIL import Image
from   sklearn.externals import joblib

from   data import MnistConfig, MnistDataset, loader
from   models import LeNet5AE, DPCCA
```

```python
path  = '/Users/gwg/dmcm/experiments/20190212_fa/model.joblib.pkl'
model = joblib.load(path)

dataset = MnistDataset(MnistConfig())
x1 = dataset.images
x2 = dataset.genes
x  = torch.cat([x1.view(-1, 28*28), x2], dim=1)
```

```python
plt.imshow(model.components_)
```

```python
z = model.transform(x)
print(z.shape)

fig, ax = plt.subplots(figsize=(22, 4))
ax.imshow(z[:400].T, interpolation='nearest')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

factor_idx = 20
z_spec     = z[:, factor_idx]
inds       = np.argsort(dataset.labels.numpy())
z_sorted   = z_spec[inds]
print(z_sorted.shape)

INT_TO_COLOR = {
    0: 'red',
    1: 'blue',
    2: 'green'
}

for i, (zi, label) in enumerate(zip(z_sorted, dataset.labels)):
    if i > 200:
        break
    color = INT_TO_COLOR[int(label.item())]
    plt.bar(i, zi, color=color)
    
plt.title('Factor %s' % factor_idx)
plt.legend(ncol=4, loc='lower center')
```
