```python
def gen_heatmap(x1_compute, gene, n_for_avg):
    heatmap = []
    names = dataset.gene_names
    gene_idx = list(names).index(gene)

    ii = 0
    for i in range(0, 980, 14):
        heatmap.append([])
        for j in range(0, 980, 14):
            x1 = x1_compute[:, i:i+28, j:j+28]
            val_avg = []
            for _ in range(n_for_avg):
                x2r = model.sample_genes_from_images(torch.Tensor(x1).unsqueeze(0))
                gene_val = x2r.squeeze()[gene_idx]
                val_avg.append(gene_val.item())
            heatmap[ii].append(np.array(val_avg).mean())
        ii += 1
        print('Done %s' % i)
    print('Heatmap completed.')
    return heatmap
```
