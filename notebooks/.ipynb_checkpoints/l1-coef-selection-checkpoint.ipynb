```python
import matplotlib.pyplot as plt
plt.rc('font', family='Arial')
plt.rc('font', size=20)
import numpy as np
import torch
from   models import DPCCA
from   data import GTExV8Config

cfg = GTExV8Config()
```

```python
def get_l1_norm(l1_coef):
    fname = '/Users/gwg/dmcm/experiments/20190115_gtex_sparse/'\
           'gtexv8_mode-dpcca_batch_size-128_seed-0_latent_dim-50_lr-1e-05_l1_coef-%s_private_z-1/'\
           'model.pt' % l1_coef
    state = torch.load(fname, map_location={'cuda:0': 'cpu'})
    model = DPCCA(cfg, latent_dim=50, private_z=True)
    model.load_state_dict(state)
    
    n_sparse = 0
    n_params = 0
    with torch.no_grad():
        l1_norm = 0
        for param in model.pcca.parameters_('y2'):
            n_params += np.prod(param.shape)
            n_sparse += (torch.abs(param) < 0.00001).sum()
            l1_norm  += torch.norm(param, 1)
    pct_sparse = n_sparse.numpy() / n_params
    return l1_norm.item(), pct_sparse
```

```python
xs  = ['0.0', '0.5', '1.0']
ys1 = []
ys2 = []

for x in xs:
    l1_norm, pct_sparse = get_l1_norm(x)
    ys1.append(l1_norm)
    ys2.append(pct_sparse)

xs_ = []
for x in xs:
    xs_.append(float(x) / (50*100))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
plt.scatter(xs_, ys1, color='red', s=200)
plt.ylabel('L1 norm')
plt.xlabel('L1 coef')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
plt.scatter(xs_, ys2, color='blue', s=200)
plt.ylabel('% sparse')
plt.xlabel('L1 coef')
```
