```python
import matplotlib.pyplot as plt
plt.rc('font', family='Arial')

import numpy as np
```

```python
fname = '/Users/gwg/dmcm/experiments/20181102_v8_big_lat_dim/'\
        'gtexv8_mode-ccaae_seed-0_latent_dim-1000_lr-1e-05_l1_coef-5.0/out.txt'

with open(fname) as out:
    lines = [l.strip() for l in out]

idx = None
for i, l in enumerate(lines):
    if l == 'Training model.':
        idx = i+2
        
losses = lines[idx:]
img_curve = []
gen_curve = []
for l in losses:
    if '=' in l:
        break

    img_curve.append(float(l.split('\t')[2]))
    gen_curve.append(float(l.split('\t')[3]))

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

x = list(range(len(img_curve)))
y = img_curve
plt.plot(x, y, color='#11accd')
y = gen_curve
plt.plot(x, y, color='#bc2612')

start, end = ax.get_ylim()
_ = ax.yaxis.set_ticks(np.arange(start, end, 180))

_ = plt.setp(ax.get_xticklabels(), fontsize=16)
_ = plt.setp(ax.get_yticklabels(), fontsize=16)
```
