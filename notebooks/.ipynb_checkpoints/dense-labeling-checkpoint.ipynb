```python
import mygene
mg = mygene.MyGeneInfo()

ens = dataset.genes_df.index.values
ens = [e.split('.')[0] for e in ens]

gene_to_ens_map = {}
for result in mg.querymany(ens, scopes='ensembl.gene'):
    if 'symbol' in result:
        gene_to_ens_map[result['symbol']] = result['query']
```

```python
print(len(gene_to_ens_map.keys()))

def gen_heatmap(x1_compute, gene, n_for_avg):
    heatmap = []
    ens = gene_to_ens_map[gene]
    all_ens = [e.split('.')[0] for e in dataset.genes_df.index.values]
    gene_idx = list(all_ens).index(ens)
    print(gene_idx)

    ii = 0
    with torch.no_grad():
        for i in range(0, 900, 50):
            heatmap.append([])
            for j in range(0, 900, 50):
                x1 = x1_whole[:, i:i+128, j:j+128] / x1_whole.max()
                if x1.shape != torch.Size([3, 128, 128]):
                    continue
                val_avg = []
                for _ in range(n_for_avg):
                    x2r = model.sample_x2_from_x1(x1.unsqueeze(0))
                    gene_val = x2r.squeeze()[gene_idx]
                    val_avg.append(gene_val.item())
                heatmap[ii].append(np.array(val_avg).mean())
            ii += 1
            print('Done %s' % i)
        print('Heatmap completed.')
    return heatmap
```

```python
heatmap = gen_heatmap(x1_whole, 'FOXE1', 3)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
heatmap = np.array(heatmap)

ax = plt.imshow(heatmap, vmin=0, vmax=1)
```
